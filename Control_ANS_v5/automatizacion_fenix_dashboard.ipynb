```python
from pathlib import Path
import win32com.client as win32
```

```python
ruta_proyecto= Path.cwd()
print("Ruta del proyecto:", ruta_proyecto)
```

```python
archivo_fenix=ruta_proyecto /"data_clean"/"FENIX_ANS.xlsx"
archivo_dashboard = ruta_proyecto /"dashboard"/"Dashboard.xlsm"
print("Archivo Fenix:", archivo_fenix)
print("Archivo Dashboard:",archivo_dashboard)
```

```python
print("¿Existe FENIX_ANS",archivo_fenix.exists())
print("¿Existe Dashboard",archivo_dashboard.exists())
```

```python
excel=win32.DispatchEx("Excel.Application")
excel.Visible=False
excel.DisplayAlerts=False

print("Excel iniciado correctamente")
```

```python
wb_fenix=excel.Workbooks.open(str(archivo_fenix))
wb_dashboard=excel.Workbooks.open(str(archivo_dashboard))
print("Archivos abiertos correctamente")
```

```python
ws_fenix=wb_fenix.sheets("FENIX_ANS")
ws_dashboard=wb_dashboard.sheets("DATOS_ANS")
print("Hoja origen:",ws_fenix.Name)
print("Hoja destino:",ws_dashboard.Name)
```

```python
ultimafila=ws_fenix.Cells(ws_fenix.Rows.Count,"A").End(-4162).Row
print("última fila detectada:",ultimafila)
```

```python
ultima_fila_dashboard = ws_dashboard.Cells(ws_dashboard.Rows.Count, "A").End(-4162).Row

if ultima_fila_dashboard > 1:
    ws_dashboard.Rows(f"2:{ultima_fila_dashboard}").Delete()
```

```python
rango_origen = ws_fenix.Range(f"A2:AF{ultimafila}")
rango_destino = ws_dashboard.Range("A2")

rango_origen.Copy(rango_destino)

print("Datos copiados correctamente")
```

```python
wb_dashboard.Save()
print("Dashboard guardado")
```

```python
wb_fenix.Close(False)
wb_dashboard.Close(True)
excel.Quit()
print("Proceso terminado")
```
